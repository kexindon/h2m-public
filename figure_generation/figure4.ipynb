```python
def find_9mers_containing_position(sequence, position):
    """
    Generate all 9-mer slices from a given DNA sequence that include a specific position.

    Args:
    sequence (str): The DNA sequence.
    position (int): The 1-based position within the sequence.

    Returns:
    list: A list of all 9-mer slices including the specified position.
    """
    # Adjust position to 0-based for easier slicing
    sequence = str(h2m.Translate(sequence))
    position = position[0]

    k_mers = [""] * 9  # 9
    # 9-mer
    start_index = position - 8

    # 9-mer
    for i in range(9):
        if 0 <= start_index + i <= len(sequence) - 9:  # 9-mer
            k_mer = sequence[start_index + i: start_index + i + 9]
            k_mers[i] = k_mer
    
    # Initialize list to collect 9-mers
    
    return k_mers
```

```python
def dict_score_class(h):
    if h <= 2:
        return 'both'
    else:
        return 'neither'
    
dict_color_class = {
    'both':'darkred',
    'neither':'grey'
}
```

```python
df_plot['class'] = [dict_score_class(h) for h in df_plot['percentile_rank']]
df_plot['color'] = [dict_color_class[x] for x in df_plot['class']]
```

```python
l_plot = df_plot
bbb1 = l_plot[(l_plot['percentile_rank'] <= 2)]
# top_2 = df_merged['HGVSp_h'].value_counts().index.to_list()[:2]
# bbb3 = df_merged[df_merged['HGVSp_h'].isin(top_2)]
bb2 = bbb1

x = list(bb2['x'])
y = list(bb2['percentile_rank'])
t = ['|'.join([x,y]) for x,y in zip(bb2['human_9mers'], bb2['allele'])]
c = list(bb2['color'])

fig, ax = plt.subplots(figsize=(6, 9), dpi=300)
ax.scatter(data=l_plot, x='x', y='percentile_rank', c = 'color',alpha=.3, s=50, edgecolor='black', linewidths=.1)
# plt.xscale('log')
plt.yscale('log')

ax.set_xlim(0, 3)
ax.set_ylim(0, 100)
ax.set_ylabel(f"Percentile Rank Human", fontsize=15)
ax.set_xlabel("", fontsize=15)

l_1 = len(l_plot.drop_duplicates(subset = 'human_9mers'))
l_2 = len(l_plot[l_plot['class']=='both'].drop_duplicates(subset = 'human_9mers'))
per=l_2/l_1*100
ax.set_title(f"{l_2}/{l_1} 9-mer peptides predicted to be\npresented by human MHC class I", fontsize=15)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='y', which='major', labelsize=15)
ax.xaxis.set_ticks([1,2])
ax.xaxis.set_ticklabels(['R521P','R521G'], size=15, fontweight=100)

texts = [ax.text(x[i],y[i],t[i], fontsize=12, color=c[i], 
         path_effects=[PathEffects.withStroke(linewidth=2, foreground="w")]) for i in range(len(x))]

adjust_text(texts,only_move={'texts':'xy'}, ax=ax, arrowprops=dict(arrowstyle="-", linewidth=1, alpha=1, color='black'))

#plt.savefig('/Users/kexindong/Documents/GitHub/Output/neoantigen/final_result/best_h2m.svg')
plt.show()
```
