```python
import bioh2m as h2m
import pandas as pd
from pathlib import Path
```

```python
df_all = pd.concat([pd.read_csv('/Users/kexindong/Documents/GitHub/Output/h2m_database/aacr_input_addon.csv'),pd.read_csv('/Users/kexindong/Documents/GitHub/Output/h2m_database/cosmic_input_addon.csv')])
df_all
df_all['ref_seq_h'] = df_all['ref_seq_h'].fillna('-')
df_all['alt_seq_h'] = df_all['alt_seq_h'].fillna('-')
```

```python
path_h_ref, path_m_ref = '/Users/kexindong/Documents/GitHub/Database/RefGenome/ncbi-2023-09-12/GCF_000001405.25_GRCh37.p13_genomic.fna.gz', '/Users/kexindong/Documents/GitHub/Database/RefGenome/mouse-2023-09-13/GCF_000001635.27_GRCm39_genomic.fna.gz'
# remember to replace the paths with yours
records_h, index_list_h = h2m.genome_loader(path_h_ref)
records_m, index_list_m  = h2m.genome_loader(path_m_ref)
path_h_anno, path_m_anno = '/Users/kexindong/Documents/GitHub/Database/Genecode/gencode_v19_GRCh37.db', '/Users/kexindong/Documents/GitHub/Database/Genecode/gencode_vm33_GRCm39.db'
# remember to replace the paths with yours
db_h, db_m = h2m.anno_loader(path_h_anno), h2m.anno_loader(path_m_anno)
```

```python
result = h2m.model_batch(df_all, records_h, index_list_h, records_m, index_list_m, db_h, db_m, 37, memory_size=1000)
# save result
result[0].to_csv(f'/Users/kexindong/Documents/GitHub/Output/h2m_database/output_addon.csv', index=False)
# save failed samples
result[1].to_csv(f'/Users/kexindong/Documents/GitHub/Output/h2m_database/leftover_addon.csv',index=False)
```
